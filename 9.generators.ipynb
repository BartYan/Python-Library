# Generators

```python
# wczesniejsza lekcja
listA = list(range(6))
listB = list(range(6))

product = []

for a in listA:
    for b in listB:
        product.append((a,b))

print(product)

product = [(a,b) for a in listA for b in listB]
print(product)

product = [(a,b) for a in listA for b in listB if a % 2 and b % 2 == 0]
print(product)

# dict 
product = {a:b for a in listA for b in listB if a % 2 and b % 2 == 0}
print(product)
```

```python
# generator 
gen = ((a,b) for a in listA for b in listB if a % 2 and b % 2 == 0)
print(gen)

print(next(gen))
print(next(gen))
print('-'*30)
for x in gen:
    print(x)
print('-'*30)

# wygenerował ju wcześniej i tu będzie pusto 
for x in gen:
    print(x)
print('-'*30)

# zdefiniuj jeszcze raz
gen = ((a,b) for a in listA for b in listB if a % 2 and b % 2 == 0)

while True:
    try:
        print(next(gen))
    except StopIteration:
        print('all values have been generated')
        break

# Generator przydaje się gdy danych jest tyle, ze niechcesz ich przechowywać w pamięci. 
# Generator zwróci wartość tylko jeden raz. Jeśli chcesz dostać te same wartości ponownie zainicjuj go ponownie.
```

## Example

```python
ports = ['WAW', 'KRK', 'GDN', 'KTW', 'WMI', 'WRO', 'POZ', 'RZE', 'SZZ',
         'LUZ', 'BZG', 'LCJ', 'SZY', 'IEG', 'RDO']
 
routes = ( (start, stop) for start in ports for stop in ports)
 
counter=0
for (start, stop) in routes:
    print("{} - {}".format(start, stop))
    counter+=1
          
print(counter)
 
##########
 
routes = ( (start, stop) for start in ports for stop in ports if start != stop)
 
counter=0
for (start, stop) in routes:
    print("{} - {}".format(start, stop))
    counter+=1
          
print(counter)
 
##########
 
routes = ( (start, stop) for start in ports for stop in ports if start < stop)
 
counter=0
for (start, stop) in routes:
    print("{} - {}".format(start, stop))
    counter+=1
          
print(counter)
```
